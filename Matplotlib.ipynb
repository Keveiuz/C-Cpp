## Plot Figure

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Resize Graph
plt.figure(figsize=(5, 3), dpi=200) # figsize=(a, b): proportion of length and width; dpi: Dots Per Inch(pixel)
                                    # size = proportion * dpi

plt.title("Basic Figure", fontdict={'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X")
plt.ylabel("Y")
#plt.xticks([0, 1, 2, 3, 4])
#plt.yticks([0, 2, 4, 6, 8])

##Line_1
x1 = [0, 1, 2, 3, 4, 5]
y1 = [0, 2, 4, 6, 8, 10]
plt.plot(x1, y1, label='y = 2x', color='#da1500', linewidth = 2, linestyle='--', marker='^', markersize=6, markeredgecolor='#da1500')
#Use shorthands notation
#fmt = '[marker][line][color]'
#plt.plot(x, y, '.-g')

##Line_2
x2 = np.arange(0, 5, 0.01)
y2 = x2**2
plt.plot(x2[:301], y2[:301], label='y = x^2', color='#fed820')
plt.plot(x2[300:], y2[300:], label='y = x^2', color='#0001da', linestyle=':')

plt.legend()
plt.savefig('Basic Figure.png', dpi = 300)
plt.show()
```

## Bar Chart


```python
labels = ['A', 'B', 'C']
values = [4, 2, 6]

bars = plt.bar(labels, values)

pattern = ['.', '/', '*']
color = ['#da1500', '#fed820', '#0001da']
for bar in bars :
    bar.set_color(color.pop(0))
for bar in bars :
    bar.set_hatch(pattern.pop(0))
    
#bars[0].set_hatch('.')
#bars[1].set_hatch('/')
#bars[2].set_hatch('*')

plt.show()
```

## Line Graph


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv('/Users/jason/Desktop/A/Brent_Oil.csv')
print(Data)
```

```python
plt.figure(figsize=(5, 6), dpi=200, facecolor='white')
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)

'''
left  = 0.125  # the left side of the subplots of the figure
right = 0.9    # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.2   # the amount of width reserved for blank space between subplots
hspace = 0.2   # the amount of height reserved for white space between subplots
'''

#Subplot_1
plt.subplot(2, 1, 1, facecolor='#3e3e3e')

plt.title("Brent Oil Fluctruation", fontdict={'fontweight':'bold', 'fontsize':'12'})

plt.plot(Data.Years, Data.Average, label = 'Average', color = '#fed820')
plt.plot(Data.Years, Data.Highest, label = 'Highest', color = '#da1500')
plt.plot(Data.Years, Data.Lowest, label = 'Lowest', color = '#0001da')

plt.xlabel('Years')
plt.ylabel('Price($USD)')

plt.xticks(Data.Years[::5])
plt.yticks([i for i in range(0, 160, 20)])
plt.legend()

#Subplot_2
plt.subplot(2, 1, 2, facecolor='#3e3e3e') #设置背景颜色
plt.title("Brent Oil Fluctruation Rate", fontdict={'fontweight':'bold', 'fontsize':'12'})
plt.plot(Data.Years, Data.Rate, label = 'Rate', color = '#108200') 

plt.xticks(Data.Years[::5])
plt.legend()

plt.show()
```

## Histogram

```python
plt.figure(figsize=(5, 3), dpi=200, facecolor='white')

plt.title('Attribution of Annual Average Oil Price', fontdict={'fontsize':16})

plt.hist(Data.Average, bins = [i for i in range(20, 140, 20)])

plt.ylabel('Frequency')
plt.xlabel('Price Pre Gallon ($USD)')

#plt.colorbar()

plt.show()
```

## Pie Chart

```python
plt.figure(figsize=(5, 3), dpi=200, facecolor='#031634')

plt.title('Median Number Proportion($60)')
High = len([i for i in Data.Average if i > 60])
Low = len([i for i in Data.Average if i <= 60])

Labels = ['Higher than 60$ pre Gallon', 'Lower than 60$ pre Gallon']
Colors = ['#e8ddcb', '#cdb380']

plt.pie([High, Low], labels = Labels, colors = Colors, autopct = '%.2f %%')

plt.show()
```

```python
plt.figure(figsize=(3, 2), dpi=300, facecolor='#ffffff')

plt.style.use('ggplot')

plt.title('Median Number Proportion($60)', fontdict={'fontweight':'bold', 'fontsize':'12'})
High = len([i for i in Data.Average if i > 100])
High_Medium = len([i for i in Data.Average if (i <= 100) & (i > 80)])
Medium = len([i for i in Data.Average if (i <= 80) & (i > 60)])
Medium_Low = len([i for i in Data.Average if (i <= 60) & (i > 40)])
Low = len([i for i in Data.Average if i <= 40])

Labels = ['Over $100', '\$80-$100', '\$60-$80', '\$40-$60', 'Under $40']
Values = [High, High_Medium, Medium, Medium_Low, Low]
Sepdis = (0, .5, 0, .2, 0)

plt.pie(Values, labels = Labels, autopct = '%.2f %%', pctdistance = 0.8, explode = Sepdis)

plt.show()
```
